```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import holoviews as hv
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# hv.extension('plotly', inline=False)
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_plot_1 = df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)

df_plot_1
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_data = StandardScaler().fit_transform(X=df_market_data)
```

```python
df_market_data.index
```

```python
# Create a DataFrame with the scaled data
scaled_market_df = pd.DataFrame(scaled_data)


# Copy the crypto names from the original data
scaled_market_df.columns = df_market_data.columns


# Set the coinid column as index
scaled_market_df.index = df_market_data.index

# Display sample data
scaled_market_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_nums = range(1,12)
```

```python
# Create an empty list to store the inertia values
inertias = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list


#setting random state of 2 for consistency 
for k in k_nums:
    k_model = KMeans(n_clusters=k, random_state=2)
    k_model.fit(scaled_market_df)
    inertias.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve


# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(data={'k':k_nums, 'inertia':inertias})
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve1 = elbow_df.hvplot.line(
    x= 'k',
    y= 'inertia',
    title='Elbow Curve',
    xticks=k_nums
)

elbow_curve1
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** It is difficult to really say, but k=3 or 4 are showing the sharpest breaks. After k=4 however, the increased value for further clusters drops significantly. And k=4 seems to be the furthest distance from a straight line drawn from the maximum and minimum points of inertia. 

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k

model_k_4 = KMeans(n_clusters=4, random_state=2)
```

```python
# Fit the K-Means model using the scaled data
model_k_4.fit(scaled_market_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4_predictions = model_k_4.predict(scaled_market_df)

# Print the resulting array of cluster values.
print(k_4_predictions)
```

```python
# Create a copy of the DataFrame
k_4_scaled_market_predictions_df = scaled_market_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
k_4_scaled_market_predictions_df['class'] = k_4_predictions

# Display sample data
k_4_scaled_market_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

k_4_scaled_market_predictions_df['coin_id'] = k_4_scaled_market_predictions_df.index

kmeans_plot1 = k_4_scaled_market_predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by='class',
    hover_cols='coin_id'
)

kmeans_plot1
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_scaled_market_df = pd.DataFrame(pca.fit_transform(scaled_market_df))
pca_scaled_market_df.columns = ['PCA1', 'PCA2', 'PCA3']

# View the first five rows of the DataFrame. 
pca_scaled_market_df.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** We sum the explained variance ratio array's three terms: ~89.5% of the variance is captured by the first three principal components of the PCA.

```python
# Create a new DataFrame with the PCA data.


# Creating a DataFrame with the PCA data
market_pca_df = pca_scaled_market_df.copy()

# Copy the crypto names from the original data
market_pca_df = market_pca_df.set_index(df_market_data.index)

# Set the coinid column as index


# Display sample data
market_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_nums = range(1,12)
```

```python
# Create an empty list to store the inertia values
inertias = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `market_pca_df`
# 3. Append the model.inertia_ to the inertia list


#setting random state of 2 for consistency 
for k in k_nums:
    k_model = KMeans(n_clusters=k, random_state=2)
    k_model.fit(market_pca_df)
    inertias.append(k_model.inertia_)

```

```python
# Create a dictionary with the data to plot the Elbow curve

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(data={'k':k_nums, 'inertia':inertias})
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve2= elbow_df.hvplot.line(
    x= 'k',
    y= 'inertia',
    title='Elbow Curve',
    xticks=k_nums
)

elbow_curve2
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** It still seems like `k=4` is the best fit k means solution when using the PCA data.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No it does not. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=2)

```

```python
# Fit the K-Means model using the PCA data
model.fit(market_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_4 = model.predict(market_pca_df)
# Print the resulting array of cluster values.
print(k_4)
```

```python
# Create a copy of the DataFrame with the PCA data
k_4_market_pca_df = market_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
k_4_market_pca_df['class'] = k_4


# Display sample data
k_4_market_pca_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
k_4_market_pca_df['coin_id'] = k_4_market_pca_df.index

kmeans_plot2 = k_4_market_pca_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by='class',
    hover_cols='coin_id'
)

kmeans_plot2
```
